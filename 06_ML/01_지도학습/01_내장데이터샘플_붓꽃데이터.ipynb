###  데이터를 쉽게 이해하기 위해 df 변경한다. 

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df["species_name"] = iris.target_names[iris.target]
df
```

### 시각화 - 산점도
* 4가지 피처들중에서 영향력이 있는 피쳐는 petal 속성이다. ==> 상관계수로 확인 가능하다.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 3가지 종류 추출

```python
df_setosa = df[df['species_name']=='setosa']
df_versicolor = df[df['species_name']=='versicolor']
df_virginica = df[df['species_name']=='virginica']
df_setosa
df_versicolor
df_virginica.head()
```

```python
plt.scatter(df_setosa.iloc[:, 2], df_setosa.iloc[:, 3])
plt.scatter(df_versicolor.iloc[:, 2], df_versicolor.iloc[:, 3])
plt.scatter(df_virginica.iloc[:, 2], df_virginica.iloc[:, 3])
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.legend(["setosa","versicolor","virginica"])
plt.show()
```
